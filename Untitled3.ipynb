# web scrapying: obtaining data from DEF 14 A

1.print the web page

```python
import requests                                                                      
from bs4 import BeautifulSoup 
import pandas as pd
import urllib
from urllib.request import urlopen
```

```python

html=urlopen("https://www.sec.gov/Archives/edgar/data/66740/000120677418000890/threem3251561-def14a.htm")
```

```python
bsObj=BeautifulSoup(html.read())
print(bsObj)
```

2.find title

```python
title=bsObj.title
print(title)
```

3.find one particular shareholder proposal

```python
p=bsObj.findAll(text="Proposal No. 1: Elect the Twelve Directors Identified in this Proxy Statement")
print(len(p))
```

```python
print(p)
```

4.collect all the stockholder proposals in the webpage

```python
x=bsObj.select("b > a")
```

```python
x
```

```python

for i in x:
       print(i.get_text())
```

```python
x[0].get_text()
```

```python
list1 = []

for i in x:
    list1.append(i.get_text())
    
```

```python
print(list1)
```

```python
list1[0]
```

```python
len(list1)
```

```python
B=[]
for each in list1:
    if 'Stockholder Proposal' in each:
        B.append(each)
print(B) 
```

```python
A=pd.DataFrame(B)
```

```python
import csv
```

```python
A.to_csv('A.csv',index= False,header= False)
```
